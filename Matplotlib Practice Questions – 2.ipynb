27. Create a multiple bar plot showing the number of hours spent on three
activities (Study, Exercise, Sleep) by three different individuals (Person A,
Person B, Person C). Use the data: Person A - [4, 1, 7], Person B - [5, 2, 6],
Person C - [3, 1, 8]. Display the bars side by side and use different colors
for each person.

```python
activities = ['Study','Exercise', 'Sleep']
A = [4, 1, 7]
B = [5, 2, 6]
C = [3, 1, 8]
# Number of categories
n = len(activities)

# X positions
x = np.arange(n)

# x = np.arange(len(categories))


# Bar widtha
width = .25

# Create bars
plt.bar(x - width, A, width, label='A', color='skyblue')
plt.bar(x, B, width, label='B', color='green')
plt.bar(x + width, C, width, label='C', color='orange')
# Add labels and title
plt.xlabel('Persons')
plt.ylabel('Values')
plt.title('Side-by-Side Bar Plot')
plt.legend(['A','B','C'])
plt.xticks(x, activities)

plt.show()
```

29. Plot a quadratic function y = -x2 + 4x  for x values from -2 to 4. Use
fill_between to shade the area under the curve. Make sure to label the axes
and add a title.

```python
x = np.linspace(-2,4,1000)
y=-x**2 + 4*x

plt.plot(x,y)
plt.fill_between(x,y, color= 'blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = -x2 + 4x')
plt.show()
```

30. Create a 2x2 grid of subplots. In the first subplot (top-left), plot a line graph of y=x^2.
In the second subplot (top-right), plot a scatter plot of x and y where y is a random
sample from a normal distribution. In the third subplot (bottom-left), plot a bar
chart of categories A, B, and C with heights 3, 7, and 2, respectively. In the fourth
subplot (bottom-right), plot a histogram of 1000 random samples from a normal
distribution.

```python
x = np.linspace(-5, 5, 100)
y = x**2

plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Line Plot')

np.random.seed(0)
y = np.random.randn(100)
plt.subplot(2, 2, 2)
plt.scatter(x, y)
plt.title('Scatter Plot')

categories = ['A', 'B', 'C']
heights = [3, 7, 2]
plt.subplot(2, 2, 3)
plt.bar(categories, heights)
plt.title('Bar Chart')

plt.subplot(2, 2, 4)
samples = np.random.randn(1000)
plt.hist(samples, bins=20)
plt.title('Histogram')

plt.tight_layout()
plt.show()
```
