11. Create a line plot with x = [0, 1, 2, 3, 4] and y = [0, 1, 4, 9, 16].
Add a grid to the plot.

13. Create a line plot with x = [0, 1, 2, 3, 4] and y = [0, 1, 4, 9, 16].
Add a title "My Plot" and label the x-axis as "Time" and the y-axis as
"Value".

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('My Plot')
plt.show()
```

14. Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [5, 4, 3, 2, 1].
Use different colors and sizes for each point.

```python
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
colors=['red','blue','green','yellow','black']
size=[100,200,300,400,500]

plt.scatter(x, y, c=colors, s = size)
plt.show()
```

17. Create a plot with x = [0, 1, 2, 3, 4] and y1 = [0, 1, 4, 9, 16], y2 =

```python
x = [0, 1, 2, 3, 4]
y1 = [0, 1, 4, 9, 16]
y2 = [0, 1, 2, 3, 4]

plt.plot(x,y1,x,y2)
plt.legend(['y1','y2'])
plt.show()
```
