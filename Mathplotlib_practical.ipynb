```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Multiple Line Plot with Different Line Styles and Widths

Plot three lines on the same graph, each with a different style and width. Use dashed, dotted, and solid line styles, and set line widths to 1.5, 2.5, and 3, respectively. Add gridlines to the plot.

```python
y1 = np.random.rand(5)
y2 = np.random.rand(5)
y3 = np.random.rand(5)

plt.plot(y1 ,'--', lw= 1.5)
plt.plot(y2 ,':', lw= 2.5)
plt.plot(y3 , lw= 3)

plt.grid()
plt.show()
```

2. Scatter Plot with Custom Marker Styles and Colors

Create a scatter plot with 10 points. Use a circle, square, and triangle marker, each with different colors and sizes. Ensure there are no connecting lines between markers.

```python
x = np.random.rand(10)
y = np.random.rand(10)

plt.scatter( x, y, marker = '*', color= 'r', s =100 ) # Use the current marker for the current point

plt.show()
```

3. Bar Graph with Edge Color and Face Color

Plot a vertical bar graph for five categories with custom colors for both the bar's face and edge. Label the x-axis as "Categories" and y-axis as "Values."

```python
cat = ['a','b','c','d','e']
x = np.random.rand(5)

plt.bar(cat,x, edgecolor= 'k', color='w')
plt.show()
```

4. Horizontal Bar Graph with Annotations

Create a horizontal bar graph for five categories. Label each bar with its respective value using annotations. Customize the face color of each bar to be different.

```python
cat = ['a', 'b', 'c', 'd', 'e']
x = np.random.randint(0,10,5)

plt.barh(cat, x, color=['b','g','r','cyan','m'])

# Get numerical positions of bars for annotation
xpos = np.arange(len(cat))

for i, j in zip(xpos, x):
    plt.annotate(f'{j}',
                 xy=(j, i),
                 xytext=(j + 0.02, i),
                 ha='left', va='center')


plt.show()
```

5. Pie Chart with Start Angle, Explode, and Shadow

Create a pie chart with four categories. Set the starting angle to 90 degrees, and explode the secon slice by 0.1. Enable the shadow effect and add labels to each slice.

```python
label = ['A','B','C','D']
value = [45,45,45,45]

plt.pie(value,labels = label, startangle = 90, explode=[0,0.1,0,0], shadow = True)
plt.show()
```

6. Subplots: Line Plot and Bar Graph

Create two subplots: a line plot in the first subplot and a bar graph in the second. Add titles for each subplot and set different colors for each plot. The line plot should include markers at each data point

```python
x = ['a', 'b', 'c', 'd', 'e']
y = [24,64,39,57,98]

plt.subplot(1,2,1)
plt.plot(x,y,'*-r')
plt.title('Line Plot')

plt.subplot(1,2,2)
plt.bar(x,y,color= 'skyblue')
plt.title('Bar Plot')

plt.show()
```

7. Line Chart with Annotations and Legend

Plot two lines on the same graph with different colors and add a legend for each. Place an annotate at the highest point of each line, marking it as "Peak Value." Display the legend outside the plot.

```python
x = ['a', 'b', 'c', 'd', 'e']
y1 = [24,64,39,57,98]
y2 = [65,81,56,45,72]

plt.plot(x,y1,'*-r')
plt.annotate('Highest Point', xy=('e',98), xytext=('c',95),
             arrowprops=dict(facecolor='red', arrowstyle='->'))


plt.plot(x,y2,'^-b')
plt.annotate('Highest Point', xy=('b',81), xytext=('c',95),
             arrowprops=dict(facecolor='red', arrowstyle='->'))

plt.title('Line Plot')
plt.show()
```

8. Fill Area Between Two Lines

Create a plot with two lines, and shade the area between them. Label the x-axis as "X-Values" an axis as "Y-Values." Use a transparent color for the fill area.

```python
x = ['a', 'b', 'c', 'd', 'e']
y1 = [24,64,39,57,98]
y2 = [65,81,56,45,72]

plt.plot(x,y1,'*-r')

plt.plot(x,y2,'^-b')

plt.fill_between( x, y1, y2, color='yellow', alpha= 0.3)
plt.title('Line Plot')
plt.show()
```

9. Multiple Bar Plot (Grouped Bar Plot)

Create a grouped bar plot for two categories (e.g., Category A and Category B) across five groups Use different colors for each category and add a legend. Label the x-axis as "Groups" and y-axis "Values."

```python
categories = ['A', 'B', 'C', 'D']
group1 = np.random.randint(1,11,4)
group2 = np.random.randint(1,11,4)

width = .4

plt.bar(x - width/2, group1, width, label='Values 1', color='skyblue')
plt.bar(x + width/2, group2, width, label='Values 2', color='orange')

plt.xlabel('Groups')
plt.ylabel('Values')
plt.legend(['Group 1','Group 2'])
plt.xticks(x,categories)
plt.show()
```

10. Line Plot with Customized Gridlines and Line Markers

Plot a line graph with a set of 10 values. Add gridlines with custom style and color. Use markers specific color and size, and set the line to a dash-dot style.

```python
y1 = np.random.randint(1,101,10)

plt.plot(y1,'*-.r',ms=10)

plt.grid(color='green', linestyle='--', linewidth=0.5)
plt.show()
```
